```python
data["DateTime"] = pd.to_datetime(data.Date_Time, format=format)
data["Mid"] = (data["High"] + data["Low"] + data["Open"] + data["Close"])/4
```

```python
data1 = data[["DateTime", "Open", "High", "Low", "Close", "Mid", "Volume"]]
filtered_data = data1[data1.DateTime>=dateTime]
```

```python
dateTime_start = "2018-10-01 09:30:00"
dateTime_end = "2018-10-01 16:00:00"
filtered_data_prime = data1[data1.DateTime>=dateTime_start]
filtered_data = filtered_data_prime[filtered_data_prime.DateTime<=dateTime_end]
```

```python
filtered_data
```

```python
### First version : NaÃ¯ve TWAP

shares = 10000
pct = 0.1 #10% of the volume

total_shares = shares
quantity_remaining = shares
trades = []
total_price = 0
n_bins = 390
interval = 1
nb_trades = np.floor(n_bins/interval)
slice_size = np.floor(1.*total_shares / (nb_trades))+1

count = 0

for row in filtered_data.itertuples():
    if (count%interval == 0):
        if (quantity_remaining < slice_size):
            trades.append({"Time" : row.DateTime, "Volume" : quantity_remaining, "Price" : row.Mid,"Volume_reference" : row.Volume})
    
            quantity_remaining -= quantity_remaining
    
            total_price += quantity_remaining * row.Mid
            
        else :
            trades.append({"Time" : row.DateTime, "Volume" : slice_size, "Price" : row.Mid,"Volume_reference" : row.Volume})
    
            quantity_remaining -= slice_size
    
            total_price += slice_size * row.Mid
    
    count += 1

    if(quantity_remaining <= 0):
        break

print("total traded price : ", total_price)
print("average traded price : ", total_price / total_shares)
data_trades = pd.DataFrame(trades)
print(pd.DataFrame(trades))
print(sum(pd.DataFrame(trades)["Volume"]))
```

```python
### Second version : Tilted TWAP

shares = 10000
prct_trigger = 0.3
fast_pace = 2

def function_tilt(x,t, tilt = 0, random = False):
    if(random):
        return(2*x*np.random.uniform(0,1))
    else:
        return(fast_pace*x + tilt*t)

total_shares = shares
quantity_remaining = shares
trades = []
total_price = 0
n_bins = 390
interval = 10
nb_trades = np.floor(n_bins/interval)
slice_size = np.floor(1.*total_shares / (nb_trades))+1

count = 0
triggered = False
tilt = 1

for row in filtered_data.itertuples():
    if (triggered):
        if (count%interval == 0):
            if (quantity_remaining < function_tilt(slice_size, t, tilt)):
                trades.append({"Time" : row.DateTime, "Volume" : quantity_remaining, "Price" : row.Mid, "Volume_reference" : row.Volume})
    
                quantity_remaining -= quantity_remaining
    
                total_price += quantity_remaining * row.Mid
        
            else : 
                trades.append({"Time" : row.DateTime, "Volume" : function_tilt(slice_size, t, tilt), "Price" : row.Mid, "Volume_reference" : row.Volume})
    
                quantity_remaining -= function_tilt(slice_size, t, tilt)
    
                total_price += function_tilt(slice_size, t, tilt)* row.Mid
            
            t+=1
    
        count += 1
    
    else :
        if (count%interval == 0):
            if (quantity_remaining < slice_size):
                trades.append({"Time" : row.DateTime, "Volume" : quantity_remaining, "Price" : row.Mid,  "Volume_reference" : row.Volume})
    
                quantity_remaining -= quantity_remaining
    
                total_price += quantity_remaining * row.Mid
        
            else :
                trades.append({"Time" : row.DateTime, "Volume" : slice_size, "Price" : row.Mid, "Volume_reference" : row.Volume})
    
                quantity_remaining -= slice_size
    
                total_price += slice_size * row.Mid
    
        count += 1
        
        if(quantity_remaining <= (1-prct_trigger)*shares):
            triggered = True
            t = 0

    if(quantity_remaining <= 0):
        break

print("total traded price : ", total_price)
print("average traded price : ", total_price / total_shares)
data_trades = pd.DataFrame(trades)
print(pd.DataFrame(trades))
print(quantity_remaining)
```
